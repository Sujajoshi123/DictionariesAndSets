```python
#Create a numpy array
#1. Create a list
#2. Convert it into a numpy array

my_list = [1,2,3,4,5]

my_numpy_array = np.array(my_list)

print(my_numpy_array)
```

```python
second_list = [6,7,8,9,10]

#Create a 2-Dimensional numpy array by nesting lists in np.array function

two_d_array = np.array([my_list, second_list])

print (two_d_array)
```

```python
#Check the total number of elements in a numpy array with .size attribute
two_d_array.size
```

```python
#Create a identity matrix 

np.identity(n = 3)
```

```python
#Create a numpy array with N rows, M columns and position of diagonal values in K
np.eye(N=3, M=5, k=0)
```

```python
#To create a numpy array filled with ones, we can use np.ones(shape = [2, 4]). this will create 
# a matrix with 2 rows and 4 columns , all elements as 1

np.ones(shape = [2,4])
```

```python
#Indexing in numpy arrays

one_d_array = np.array([1,2,3,4,5,6])

one_d_array[0]
```

```python
one_d_array[:4]
```

```python
#Create a two d array

two_d_arr = np.array([one_d_array , one_d_array + 3, one_d_array + 6])

print (two_d_arr)
```

```python
#Performing indexing and slicing on a two d array
#Print the element from 1st row and 4th column

two_d_arr[1,4] 
```

```python
two_d_arr[1:, 4:]
```

```python
two_d_arr[::-1, ::-1]
```

```python
#Reshaping an array , here we convert the above 3 * 6 array into 6 * 3 array

reshaped_arr = np.reshape(two_d_arr , newshape=(6,3))
print (reshaped_arr)
```

```python
#Concatenate two arrays
array_to_join = ([[10,20,30] , [40,50,60] , [70,80,90]])

np.concatenate((two_d_arr, array_to_join), axis = 1)
```
